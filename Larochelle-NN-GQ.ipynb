```python
def loss(f_x,y):
    if y.shape[0] > 1:
        i = np.where(y == 1)[0][0]
        return -np.log(f_x[i]) # negative log-likelihood
    else:
        return -np.log(f_x)
#print f_x, y
#print
#print -np.log(f_x)
#print
#return -np.log(f_x)
#print f_x[1], y[1]
#return -np.log(f_x[1])

def forward_prop(x, W1, b1, W2, b2, W3, b3):
    def sigm(z):
        return 1/(1+np.exp(-z))
    
    def tanh(z):
        return (np.exp(2*z)-1)/(np.exp(2*z)+1)

    def softmax(z):
        return np.exp(z)/np.sum(np.exp(z))

    z1 = b1 + np.dot(W1,x)
    a1 = sigm(z1)

    z2 = b2 + np.dot(W2,a1)
    a2 = sigm(z2)

    z3 = b3 + np.dot(W3,a2)
    a3 = softmax(z3)

    f_x = a3
    return z1, a1, z2, a2, z3, a3, f_x

# testing
print X_test.shape
testing_losses = []
for x,y in zip(X_test, y_test):

    x = scaler.transform(x)
    x = x.reshape(x.shape[0],1)
    y = y.reshape(y.shape[0],1)

    z1, a1, z2, a2, z3, a3, f_x = forward_prop(x, W[0], b[0], W[1], b[1], W[2], b[2])
    
    if y[0][0] == 0:
        print y[0],y[1]
        print f_x[0],f_x[1]
    
    testing_loss = np.round(loss(f_x,y),2)
    testing_losses.append(testing_loss)

plt.title('TESTING LOSSES')
plt.plot(testing_losses)
plt.legend(['Testing'])
plt.xlabel('Example')
plt.ylabel('Loss')
plt.show()
```
